```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv", index_col=[0]).reset_index(drop=True)
def clean(df):
    columns = []
    for name in df.columns:
        name = name.lower().replace(" ", "_")
        columns.append(name)
    df.columns = columns
    return df

# Clean the column names
df = clean(df)
display(df.head())

# Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 
# and have a response of "Yes" to the last marketing campaign.
df1 = df.query('total_claim_amount > 1000 & response == "Yes"')
display(df1.head())
```

```python
# Using the original Dataframe, analyze the average total_claim_amount by each policy type and 
# gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.
df2 = df.query('response=="Yes"')
display(df2.head())
df2 = df2.pivot_table(index=['policy_type', 'gender'], values=['total_claim_amount'], aggfunc='sum')
display(df2)
# Conclusions:
# Personal Auto owners have a higher claim amount than corporate and special auto
# Females in all groups have a higher claim amount, but not by large margin
```

```python
# 3. Analyze the total number of customers who have policies in each state,
# and then filter the results to only include states where there are more than 500 customers.
display(df.head())
df3 = df.query('number_of_policies > 0')
df3 = df3.pivot_table(index=['state'], values=['customer'], aggfunc='count')
df3 = df3.query('customer > 500')
display(df3)
```

```python
# 4. Find the maximum, minimum, and median customer lifetime value by education level 
# and gender. Write your conclusions.
display(df.head())
df4 = df.pivot_table(index=['education', 'gender'], values=['customer_lifetime_value'], aggfunc=['max', 'min', 'median'])
df4 = df4.sort_values(by=[('median', 'customer_lifetime_value')], ascending=False)
display(df4)
# The lower the education the higher the expected life time value, male tends to be higher than female, higher education genders are more careful and older
# The max is quite higher for lower education genders
```
